```python
# 데이터 불러오기
import numpy as np
import pandas as pd

X_train = pd.read_csv('/content/data/mpg_X_train.csv', encoding='utf-8')
X_test = pd.read_csv('/content/data/mpg_X_test.csv', encoding='utf-8')


display(X_train.head())
display(X_train.shape)   # (278,8)
display(X_train.isna().sum())

display(X_test.head())
display(X_test.shape)   # (120,8)
display(X_test.isna().sum())
```

```python
# horsepower에서 발생하는 결측치를 평균으로 대체
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean')

# 결측치를 대체하면서, numeric만 가져옴
X_train_num = imputer.fit_transform(X_train.drop('name', axis=1, inplace=False))
X_test_num = imputer.fit_transform(X_test.drop('name', axis=1, inplace=False))

# numpy.ndarray를 pandas.DataFrame으로 변환
X_train_num_df = pd.DataFrame(X_train_num, columns=X_train.drop('name', axis=1).columns)
X_test_num_df = pd.DataFrame(X_test_num, columns=X_test.drop('name', axis=1).columns)

# 결과 확인
display(X_train_num_df.isna().sum())
display(X_test_num_df.isna().sum())
```

# 수치형 변수 처리 => 데이터 스케일링

```python
## Stadard Scaler : 평균이 0, 분산이 1인 정규분포로 스케일링함

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Train 데이터의 fitting과 스케일링
scaler.fit(X_train_num_df)
X_train_sc = scaler.transform(X_train_num_df)

# Test 데이터의 fitting과 스케일링
X_test_sc = scaler.transform(X_test_num_df)

# 결과 확인
result = {
    'X_train_sc' : [X_train_sc.min(), X_train_sc.max(),
                    X_train_sc.mean(), X_train_sc.std()],
    'X_test_sc'  : [X_test_sc.min(), X_test_sc.max(),
                    X_test_sc.mean(), X_test_sc.std()]
}
result_df = pd.DataFrame(result,
                         index = ['min','max', 'mean', 'std'])
display(round(result_df, 2))
```

```python
## Min-Max Scaler : 최소값을 0, 최대값을 1로 하여 스케일링함

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Train 데이터의 fitting과 스케일링
scaler.fit(X_train_num_df)
X_train_sc = scaler.transform(X_train_num_df)

# Test 데이터의 fitting과 스케일링
X_test_sc = scaler.transform(X_test_num_df)

# 결과 확인
result = {
    'X_train_sc' : [X_train_sc.min(), X_train_sc.max(),
                    X_train_sc.mean(), X_train_sc.std()],
    'X_test_sc'  : [X_test_sc.min(), X_test_sc.max(),
                    X_test_sc.mean(), X_test_sc.std()]
}
result_df = pd.DataFrame(result,
                         index = ['min','max', 'mean', 'std'])
display(round(result_df, 2))
```

```python
# Max Abs Scaler : 최소절대값을 0, 최대절대값을 1로하여 스케일링함

from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()

# Train 데이터의 fitting과 스케일링
scaler.fit(X_train_num_df)
X_train_sc = scaler.transform(X_train_num_df)

# Test 데이터의 fitting과 스케일링
X_test_sc = scaler.transform(X_test_num_df)

# 결과 확인
result = {
    'X_train_sc' : [X_train_sc.min(), X_train_sc.max(),
                    X_train_sc.mean(), X_train_sc.std()],
    'X_test_sc'  : [X_test_sc.min(), X_test_sc.max(),
                    X_test_sc.mean(), X_test_sc.std()]
}
result_df = pd.DataFrame(result,
                         index = ['min','max', 'mean', 'std'])
display(round(result_df, 2))
```

```python
# Robust Scaler : 평균과 분산 대신 중앙값(Median)과 사분위 값을 활용하는 방식
# 중앙값을 0으로 설정하고 IQR을 사용하여 이상치의 영향을 최소하한다.
# quantile_range 파라미터(default[0.25, 0.75])를 조정하여 더 넓거나
# 좁은 범위의 값을 이성치로 설정하여 정제할 수 있다.
from sklearn.preprocessing import RobustScaler

# quantile_range를 (10.0, 90.0)로 설정
#scaler = RobustScaler(quantile_range=(10.0, 90.0))
scaler = RobustScaler()

# Train 데이터의 fitting과 스케일링
scaler.fit(X_train_num_df)
X_train_sc = scaler.transform(X_train_num_df)

# Test 데이터의 fitting과 스케일링
X_test_sc = scaler.transform(X_test_num_df)

# 결과 확인
result = {
    'X_train_sc' : [X_train_sc.min(), X_train_sc.max(),
                    X_train_sc.mean(), X_train_sc.std()],
    'X_test_sc'  : [X_test_sc.min(), X_test_sc.max(),
                    X_test_sc.mean(), X_test_sc.std()]
}
result_df = pd.DataFrame(result,
                         index = ['min','max', 'mean', 'std'])
display(round(result_df, 2))
```

```python
# numpy.ndarray를 pandas.DataFrame으로 변환
X_train_sc_df = pd.DataFrame(X_train_sc,
                             columns=X_train.drop('name', axis=1).columns)
display(X_train_sc_df.head())

# 원본 스케일로 변경하기
X_train_original = scaler.inverse_transform(X_train_sc_df)

X_train_original_df = pd.DataFrame(X_train_original,
                             columns=X_train.drop('name', axis=1).columns)

display(X_train_original_df.head())
```
