```python
iris.loc[1, 'sepal length (cm)'] = 5
iris.loc[1:3, 'sepal length (cm)']
```

```python
new_iris = pd.DataFrame({'sepal length (cm)' : [4.9],
                         'sepal width (cm)' : [3.0],
                         'petal length (cm)' : [1.4],
                         'petal width (cm)' : [0.2]})
iris = pd.concat([iris, new_iris], ignore_index=True)
print(iris.iloc[-1]) # print(iris.tail(1))
```

```python
print(iris.head())
```

```python
iris['length_total (cm)'] = iris['sepal length (cm)'] + iris['petal length (cm)']
iris['width_total (cm)'] = iris['sepal width (cm)'] + iris['petal width (cm)']

from tabulate import tabulate
print(tabulate(iris.head(), headers='keys', tablefmt='grid'))
```

```python
iris.drop(columns=['length_total (cm)', 'width_total (cm)'], inplace = True)
print(tabulate(iris.head(), headers='keys', tablefmt='grid'))
```

```python
import pandas as pd

sepal_length_mean = iris['sepal length (cm)'].mean()
print(sepal_length_mean)
sepal_length_above_mean = iris[iris['sepal length (cm)'] >= sepal_length_mean]
print(len(sepal_length_above_mean))
print(tabulate(sepal_length_above_mean.head(), headers="keys", tablefmt="grid"))
```

```python
sepal_length_std = iris['sepal length (cm)'].std()
print(sepal_length_std)
threshold = 2
sepal_length_outlier = iris[(iris['sepal length (cm)'] >= sepal_length_mean + (threshold * sepal_length_std)) |
                            (iris['sepal length (cm)'] <= sepal_length_mean - (threshold * sepal_length_std))]
print(len(sepal_length_outlier))
print(tabulate(sepal_length_outlier, headers="keys", tablefmt="grid"))
```
