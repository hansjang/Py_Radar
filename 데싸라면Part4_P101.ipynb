```python
# 데이터 불러오기
import numpy as np
import pandas as pd

df = pd.read_csv('/content/data/greats200.csv', encoding='utf-8')

display(df.head())
display(df.shape)   # (200,11)
display(df.isna().sum())
```

```python
# 결측치 행 확인
missing_rows = df[df.isna().any(axis=1)]
display(missing_rows) # [7,12, 25, 30, 35, 53]

# 결측치 ffill()로 대체
df2 = df.ffill()

display(df2.iloc[[7,12, 25, 30, 35, 53]])
display(df2.isna().sum())

# 적정값 대체
df2.loc[7,'DEATH_YEAR'] = 474
display(df2.iloc[[7]])
```

```python
# 종속변수(y) = 'SONG100' 으로 해서 데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df2.drop(columns='SONG100'),
                                                             df2['SONG100'],
                                                             test_size = 0.2,
                                                             random_state = 1004)
print('X_train :', X_train.shape, 'X_test :', X_test.shape)
print('y_train :', y_train.shape, 'y_test :', y_test.shape)
```
