```python
import numpy as np
import pandas as pd
house = pd.read_csv('/content/data/kc_house_data.csv')
# display(house.head())
display(house.info())
house = house[['price', 'sqft_living']]
## 독립변수와 종속변수의 선형 가정
house.corr()
```

```python
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
# 변수 할당
y = house['price']
x = house['sqft_living']
# 단순 선형 회귀 모형 적합
lr = ols('price ~ sqft_living', data = house).fit()
print(lr.summary())
```

```python
# 경사하강법
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np

# 데이터 준비
house = pd.read_csv('/content/data/kc_house_data.csv')
house = house[['price', 'sqft_living']]
y = house['price']
x = house['sqft_living']

x = np.array(x).reshape(21613, 1)
y = np.array(y).reshape(21613, 1)

# 데이터 분리 (학습 및 테스트)
X_train, X_test, y_train, y_test = train_test_split(x, y.ravel(), test_size=0.2)

# SGDRegressor 모델 훈련
model = SGDRegressor(max_iter=1000, tol=1e-3)
model.fit(X_train, y_train)

# 예측
y_pred = model.predict(X_test)

# 성능 평가
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R² Score: {r2}")
```

```python
import statsmodels.api as sm
import pandas as pd

# 데이터 로드
X = cars[['EngineSize', 'RPM', 'Weight', 'Length',
          'MPGcity', 'MPGhighway']]
y = cars['Price']

def stepwise_selection(X, y, threshold_in=0.05, threshold_out=0.05):
    initial_features = []
    remaining_features = list(X.columns)
    selected_features = []

    while remaining_features or selected_features:
        changed = False

        # 전진 선택 (Forward Selection)
        best_pval = threshold_in
        best_feature = None
        for feature in remaining_features:
            model = sm.OLS(y, sm.add_constant(X[selected_features + [feature]])).fit()
            pval = model.pvalues[feature]
            if pval < best_pval:
                best_pval = pval
                best_feature = feature

        if best_feature is not None:
            remaining_features.remove(best_feature)
            selected_features.append(best_feature)
            changed = True
            print(f'Added {best_feature}, p-value: {best_pval}')

        # 후진 소거 (Backward Elimination)
        model = sm.OLS(y, sm.add_constant(X[selected_features])).fit()
        pvals = model.pvalues[1:]  # 첫번째 항은 절편이므로 제외
        worst_pval = pvals.max()
        if worst_pval > threshold_out:
            worst_feature = pvals.idxmax()
            selected_features.remove(worst_feature)
            remaining_features.append(worst_feature)
            changed = True
            print(f'Removed {worst_feature}, p-value: {worst_pval}')

        if not changed:
            break

    return selected_features

selected_features = stepwise_selection(X, y)

print("Selected Features:", selected_features)
# Added Weight, p-value: 2.395384226412229e-12
# Added RPM, p-value: 9.859509192238164e-05
# Added EngineSize, p-value: 0.0016325477277142897
# Selected Features: ['Weight', 'RPM', 'EngineSize']
```
