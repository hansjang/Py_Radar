```python
import pandas as pd
from tabulate import tabulate

students = pd.DataFrame({ '이름' : ['임꺾정', '홍길동', '신사임당',
                                  '이순신', '장영실', '정약용'],
                          '국어' : [70, 85, None, 100, None, 88],
                          '영어' : [65, 100, 80, 95, None, 70],
                          '수학' : [50, 90, None, None, None, 94]})

print(tabulate(students, headers='keys', tablefmt='grid'))
print(students.isna().sum())    # 열 합계
print(students.isna().sum(1))   # 행 합계
```

```python
students1 = students.dropna()   # 결측값이 하나라도 있으면 행 삭제
print(tabulate(students1, headers='keys', tablefmt='grid'))
```

```python
students2 = students.dropna(thresh=2) # 결측값이 아닌 값이 2개이상 있으면 행 유지
print(tabulate(students2, headers='keys', tablefmt='grid'))
```

```python
students3 = students.dropna(subset=['수학'])  # 수학 과목 결측값이 있는 행만 삭제
print(tabulate(students3, headers='keys', tablefmt='grid'))
```

```python
import pandas as pd
from tabulate import tabulate

# 예제 데이터 만들기
students = pd.DataFrame({ '이름' : ['임꺾정', '홍길동', '신사임당', '이순신', '장영실',
                                   '정약용', '허난설헌', '김구', '안중근', '윤봉길'],
                          '국어' : [70, 85, None, 100, None, 88, 79, 65, 90, None],
                          '영어' : [65, 100, 80, 95, None, 70, 90, 95, 100, 70],
                          '수학' : [50, 90, None, None, None, 94, 89, 62, 43, None]})

students = students.fillna(round(students.mean(numeric_only=True), 0))
students['총점'] = students['국어'] + students['영어'] + students['수학']
print(tabulate(students, headers="keys", tablefmt="grid"))
```

```python
# 인덱스 기준으로 내림차순 정렬
students.sort_index(ascending=False, inplace=True)
print(tabulate(students, headers="keys", tablefmt="grid"))
```

```python
# 컬럼명 기준으로 오름차순 정렬(가나다순)
students.sort_index(axis=1, ascending=True, inplace=True)
print(tabulate(students, headers="keys", tablefmt="grid"))
```

```python
# 원하는 순서로 컬럼 재배열
students = students[['이름', '국어', '영어', '수학', '총점']].reset_index(drop=True)
print(tabulate(students, headers="keys", tablefmt="grid"))
```

```python
# 총점 기준으로 내림차순 정렬
# students.sort_values(by='총점', ascending=False, inplace=False).reset_index(drop=True)
students  = students.sort_values(by='총점', ascending=False, inplace=False).reset_index(drop=True)

print(tabulate(students, headers="keys", tablefmt="grid"))
```

```python
# 새로운 학생을 Series로 생성
new_student = pd.Series(['박연', 60, 50, 40, 150], index=students.columns)
# 새로운 학생을 DataFrame에 추가
students = pd.concat([students, new_student.to_frame().T], ignore_index=True)
print(tabulate(students, headers="keys", tablefmt="grid"))
```

```python
# 새로운 과목 '미술' 추가
arts = pd.DataFrame({'미술' : [80, 70, 90, 100, 85, 75, 69, 78, 88, 91, 100]})
students = pd.concat([students, arts], axis=1)

# '미술'을 수학과 총점 사이에 추가
students = students[['이름', '국어', '영어', '수학', '미술', '총점']].reset_index(drop=True)

# 총점 재계산
students['총점'] = students[['국어', '영어', '수학', '미술']].sum(axis=1)

print(tabulate(students, headers="keys", tablefmt="grid"))

# '미술' 컬럼을 삭제하고 싶다면?
students.drop(columns='미술', inplace=True)

# 총점 재계산
students['총점'] = students[['국어', '영어', '수학']].sum(axis=1)


print(tabulate(students, headers="keys", tablefmt="grid"))
```

```python
students = pd.DataFrame({ '이름' : ['임꺾정', '홍길동', '신사임당', '이순신', '장영실',
                                   '정약용', '허난설헌', '김구', '안중근', '윤봉길'],
                          '국어' : [70, 85, 82, 100, 100, 88, 79, 65, 90, 78],
                          '영어' : [65, 100, 80, 95, 82, 70, 90, 95, 100, 70],
                          '수학' : [50, 90, 71, 71, 71, 94, 89, 62, 43, 71]})
students['총점'] = students[['국어', '영어', '수학']].sum(axis=1)

print(tabulate(students, headers='keys', tablefmt='grid'))
```

```python
# 조인할 예제 데이터 만들기
students2 = pd.DataFrame({ '이름' : ['신사임당', '허난설헌', '유관순', '윤심덕', '윤봉길'],
                          '국어' : [82, 79, 85, 99, 100],
                          '영어' : [80, 90, 65, 66, 100],
                          '수학' : [71, 89, 72, 33, 100]})
students2['총점'] = students2[['국어', '영어', '수학']].sum(axis=1)

print(tabulate(students2, headers="keys", tablefmt="grid"))
```

```python
# Inner Join
students_join = pd.merge(students, students2, on='이름', how='inner')

print(tabulate(students_join, headers="keys", tablefmt="grid"))
```

```python
# Outer Join
students_join = pd.merge(students, students2, on='이름', how='outer')

print(tabulate(students_join, headers="keys", tablefmt="grid"))
```

```python
# 과목별로 점수가 같을 경우 하나로 통합 / 점수가 다를 경우 기준이 되는 x값 출력
for subject in ['국어', '영어', '수학', '총점']:
    students_join[subject] = students_join[f'{subject}_x'].combine_first(students_join[f'{subject}_y'])

# 필요한 컬럼만 유지
students_join = students_join[['이름', '국어', '영어', '수학', '총점']]

print(tabulate(students_join, headers="keys", tablefmt="grid"))
```

```python
# 과목별로 점수가 같을 경우 하나로 통합 / 점수가 다를 경우 대상이 되는 y값 출력
for subject in ['국어', '영어', '수학', '총점']:
    students_join[subject] = students_join[f'{subject}_y'].combine_first(students_join[f'{subject}_x'])

# 필요한 컬럼만 유지
students_join = students_join[['이름', '국어', '영어', '수학', '총점']]

print(tabulate(students_join, headers="keys", tablefmt="grid"))
```

```python
# 과목별로 점수가 다를 경우 평균으로 나타내기
for subject in ['국어', '영어', '수학', '총점']:
    students_join[subject] = students_join[[f'{subject}_x', f'{subject}_y']].mean(axis=1)

# 필요한 컬럼만 유지
students_join = students_join[['이름', '국어', '영어', '수학', '총점']]

print(tabulate(students_join, headers="keys", tablefmt="grid"))
```

```python
# students_join = students_join.drop('총점', axis=1)
students_join['총점'] = students_join[['국어', '영어', '수학']].sum(axis=1)

# 총점순 정렬
students_join = students_join.sort_values(by='총점', ascending=False, inplace=False).reset_index(drop=True)

print(tabulate(students_join, headers="keys", tablefmt="grid"))
```

```python
import pandas as pd
from tabulate import tabulate

# students 데이터프레임 재생성
students = pd.DataFrame({
    '이름': ['홍길동', '이순신', '윤봉길', '허난설헌', '장영실', '정약용',
           '신사임당', '안중근', '김구', '유관순', '윤심덕', '임꺾정'],
    '시대': ['조선', '조선', '일제강점기', '조선', '조선', '조선', '조선',
           '일제강점기', '일제강점기', '일제강점기', '일제강점기', '조선'],
    '국어': [85, 100, 89, 79, 100, 88, 82, 90, 65, 85, 99, 70],
    '영어': [100, 95, 85, 90, 82, 70, 80, 100, 95, 65, 66, 65],
    '수학': [90, 71, 85.5, 89, 71, 94, 71, 43, 62, 72, 33, 50],
    '총점': [275, 266, 259.5, 258, 253, 252, 233, 233, 222, 222, 198, 185]
})

# 시대별 국어, 영어, 수학, 총점 평균 계산
grouped_students = students.groupby('시대')[['국어', '영어', '수학', '총점']].mean().reset_index()

# 원하는 순서로 인덱스를 재정렬
grouped_students = grouped_students.iloc[[1, 0]].reset_index(drop=True)


# 결과 출력
print(tabulate(grouped_students, headers="keys", tablefmt="grid"))
```

```python
# 총점 구간 설정
bins = [0, 200, 210, 220, 230, 240, 250, 260, 270, 280]  # 200미만, 200 이상부터는 10단위
labels = ['<200', '200-209', '210-219', '220-229', '230-239', '240-249', '250-259', '260-269', '270-279']

# 히스토그램 데이터 생성
students['총점 구간'] = pd.cut(students['총점'], bins=bins, labels=labels, right=False)

# 구간별 데이터 수 계산
frequency_distribution = students['총점 구간'].value_counts().sort_index()

# 도수분포표 데이터프레임 생성
frequency_df = pd.DataFrame({
    '총점 구간': frequency_distribution.index,
    '도수': frequency_distribution.values
})

# 결과 출력
print(tabulate(frequency_df, headers='keys', tablefmt='grid'))
```

```python
import matplotlib.pyplot as plt

# 히스토그램 시각화
plt.figure(figsize=(10, 6))
frequency_distribution.plot(kind='bar', color='skyblue')
plt.xlabel('Total Score Range')
plt.ylabel('Students')
plt.title('Number of Students by Score Range')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# 결과 출력
plt.show()
```
