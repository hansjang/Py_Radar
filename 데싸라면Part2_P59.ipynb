```python
# 생년월일, 사망년월일 날짜 형식으로 변환하기
# errors='coerce' : 날짜 형식으로 변환할 수 없는 값은 NaT(Not a Time)으로 처리
# format='mixed' : 날짜 형식이 다양한 경우, 각 값에 대해 형식을 자동으로 추론
df['생년월일'] = pd.to_datetime(df['생년월일'], errors='coerce', format='mixed')
df['사망년월일'] = pd.to_datetime(df['사망년월일'], errors='coerce', format='mixed')

# 잘 변환되었는지 확인
print(tabulate(df, headers="keys", tablefmt="grid"))
```

```python
# ['생애년'] 추가
df['생애년'] = (df['사망년월일'] - df['생년월일']).dt.days // 365.25

# 생애년을 정수형으로 변환
df['생애년'] = df['생애년'].astype(int)

# 생애년 내림차순 정렬
df = df.sort_values(by='생애년', ascending=False, inplace=False).reset_index(drop=True)

# 원하는 순서로 컬럼 재배열
df = df[['이름', '시대', '생애년', '생년월일', '사망년월일',
         '진로', '국어', '영어', '수학', '사회', '과학', '총점']].reset_index(drop=True)

# 결과 확인
print(tabulate(df, headers='keys', tablefmt='grid'))

# 생애년 평균 구하기
print(f'생애년 평균값=', round(df['생애년'].mean(),1), '세')
print(f'생애년 중간값=', round(df['생애년'].median(),1), '세')
```
