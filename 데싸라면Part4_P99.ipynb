```python
# 데이터 불러오기
import numpy as np
import pandas as pd

df = pd.read_csv('/content/data/Greats47.csv', encoding='utf-8')

display(df.head())
display(df.shape)   # (47,11)
```

# 범주형변수 처리 => 라벨인코딩, 원핫인코딩

```python
# 순서형 변수의 라벨 인코딩(Label Encoding)
from sklearn.preprocessing import LabelEncoder

# Label Encoder 적용
encoder = LabelEncoder()
df['AGE_encoded'] = encoder.fit_transform(df['AGE'].values.ravel()) # 1차원 배열로 전달

display(df[['AGE', 'AGE_encoded']])  # 조선후기 = 2, 일제강점기 = 1, 대한민국 = 0
```

```python
# 순서를 유지한 인코딩(Ordinal Encoding)
from sklearn.preprocessing import OrdinalEncoder

# Ordinal Encoder 적용
encoder = OrdinalEncoder(categories=[['조선후기', '일제강점기', '대한민국']])
df['AGE_encoded'] = encoder.fit_transform(df[['AGE']])
df['AGE_encoded'] = df['AGE_encoded'].astype(int)   # float로 반환되기 때문에 int로 변환

display(df[['AGE','AGE_encoded']]) # 조선후기 = 0, 일제강점기 = 1, 대한민국 = 2
```

```python
# 범주형 변수의 더미변수화 (pd.get_dummies)
import pandas as pd

df_encoded = pd.get_dummies(df, columns=['AGE'])

display(df_encoded.head())  # 새로운 변수 AGE_대한민국, AGE_일제강점기, AGE_조선후기가 생성됨
                            # 변수값은 True, False로 반환됨
```

```python
# 범주형 변수의 더미변수화 (원핫인코더)
from sklearn.preprocessing import OneHotEncoder

# One-Hot Encoder 적용
encoder = OneHotEncoder(sparse_output=False)
arr_encoded = encoder.fit_transform(df[['AGE']]) # df를 받아서 ndarray로 반환함

# 결과를 데이터 프레임으로 변환
df_encoded = pd.DataFrame(arr_encoded,
                          columns=encoder.get_feature_names_out(['AGE']))
# 새로운 변수 AGE_대한민국, AGE_일제강점기, AGE_조선후기가 생성됨
# 변수값은 1 (True), 0 (False)로 반환됨

# display(df_encoded.head())

# 원본 데이터 프레임과 결합
df = pd.concat([df, df_encoded], axis=1)     # 결합
df1 = df.drop('AGE', axis=1, inplace=False)  # 변환전 컬럼 'AGE' 삭제
                                             # inplace=False로 원본 유지
display(df1.head())
```
